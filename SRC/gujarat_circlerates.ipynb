```python
merge = []
ctr = 0
for x in df:
    if len(x.df)<3:
        continue
    raw_tab = x.df
    tab = raw_tab[4:].copy()
    tab.rename(columns={0:"SurveyNo",
                       1:"CircleRate1",
                       2:"CircleRate2",
                       3:"CircleRate3",
                       4:"CircleRate4",
                       5:"Col5Notes"}, inplace=True)
    tab.CircleRate1 = tab.CircleRate1.str.split("/").str[0]
    tab.CircleRate2 = tab.CircleRate2.str.split("/").str[0]
    merge.append(tab)
    print(f"{ctr}", end="\r")
    ctr+=1
```
